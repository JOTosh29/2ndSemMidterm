```python
#1
import csv
import pandas as pd
import hvplot.pandas


data = pd.read_csv(r'C:\Users\Jim Owen\Downloads\archive\usedcarsdata.csv')

cars_per_fuel_type = data['Fuel_Type'].value_counts()
cars_per_fuel_type.hvplot.bar(title='Total Cars per Fuel Type', xlabel='Fuel Type', ylabel='Number of Cars')
```

```python
#2
avg_price_by_year = data.groupby('Year')['Price'].mean()
avg_price_by_year.hvplot.line(title='Average Price of Cars by Year', xlabel='Year', ylabel='Average Price')
```

```python
#3
data.hvplot.scatter(x='Kilometers_Driven', y='Price', title='Kilometers Driven vs. Price', xlabel='Kilometers Driven', ylabel='Price')
```

```python
#4
data['Year'].hvplot.hist(title='Distribution of Car Manufacturing Years', xlabel='Year', ylabel='Frequency')
```

```python
#5
data.hvplot.barh(y='Price', x='Fuel_Type', title='Price Distribution by Fuel Type', ylabel='Price', xlabel='Fuel Type', line_color='none')
```

```python
#6
cars_per_location_year = data.groupby(['Location', 'Year']).size().unstack().fillna(0)
cars_per_location_year.hvplot.bar(stacked=True, xlabel='Location', ylabel='Number of Cars', title='Total Cars per Location and Year')
```

```python
#7
avg_mileage_by_year = data.groupby('Year')['Kilometers_Driven'].mean()
avg_mileage_by_year.hvplot.line(title='Average Mileage of Cars by Year', xlabel='Year', ylabel='Average Mileage',y= 'Kilometers_Driven')
```

```python
#8
owner_type_distribution = data['Owner_Type'].value_counts()
owner_type_distribution.hvplot.bar(title='Owner Type Distribution', xlabel='Owner Type', ylabel='Number of Cars')
```

```python
#9
data.groupby(['Location', 'Year']).size().unstack().hvplot.line(title='Total Cars per Location over Years')
```

```python
#10
data.hvplot.barh(y='Price', x='Fuel_Type', title='Price Distribution by Fuel Type',
                        ylabel='Price', xlabel='Fuel Type', line_color='none')
```

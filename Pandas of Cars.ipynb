```python
#1
import csv
import pandas as pd


data = pd.read_csv(r'C:\Users\Jim Owen\Downloads\archive\usedcarsdata.csv')
columns = ['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type']

display(data[columns])
```

```python
#2
data_mumbai = data[data['Location'] == 'Mumbai'][['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type']]
display(data_mumbai)
```

```python
#3
data_2015 = data[data['Year'] == 2015][['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type']]
display(data_2015)
```

```python
#4
highest_mileage_car = data[data['Kilometers_Driven'] == data['Kilometers_Driven'].max()][['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type']]
display(highest_mileage_car)
```

```python
#5
fuel_type_counts = data['Fuel_Type'].value_counts().reset_index().rename(columns={'index': 'Fuel_Type', 'Fuel_Type': 'Count'})
display(fuel_type_counts.sort_values(by='count', ascending=False))
```

```python
#6
avg_price_transmission = data.groupby('Transmission')['Price'].mean().reset_index().sort_values(by='Price', ascending=False)
display(avg_price_transmission)
```

```python
#7
seats_owner_type = data.groupby('Owner_Type')['Seats'].sum().reset_index().sort_values(by='Seats', ascending=False)
display(seats_owner_type)
```

```python
#8
cars_per_location_year = data.groupby(['Location', 'Year']).size().reset_index(name='Count').sort_values(by='Count', ascending=False)
display(cars_per_location_year)
```

```python
#9
cars_in_mumbai = data[data['Location'] == 'Mumbai'].groupby(['Year', 'Fuel_Type', 'Owner_Type']).size().reset_index(name='Count')
display(cars_in_mumbai)
```

```python
#10
mumbai_max_cars = data[data['Location'] == 'Mumbai'].groupby('Year').size().reset_index(name='Count')
mumbai_max_cars[mumbai_max_cars['Count'] == mumbai_max_cars['Count'].max()][['Year', 'Count']]
display(mumbai_max_cars)
```
